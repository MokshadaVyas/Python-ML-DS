```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as linear_model
```

```python
df = pd.read_csv("Home_price.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('price')
plt.scatter(df.Area , df.Price , color='red' , marker = "+")
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df[['Area']], df.Price)  # Make sure 'price' is accessed properly
```

```python
prediction = reg.predict(pd.DataFrame({'Area': [3300]}))
print(prediction)
```
